```python
Top_Teams["Penalty_S_R"] = ((Top_Teams.penalty_save / Top_Teams.penalty_conceded) * 100).astype(int)
```

```python
#VISUALIZATION
Labels = ["Man_U", "Man_C", "CHE", "LIV", "ARS", "TOT"]
figure, Axes = plt.subplots(figsize = (14,7))
Axes.bar(Top_Teams["season"], Top_Teams["wins"])
Axes.set_ylabel = ("season")
Axes.set_xlabel = ("wins count")
Axes.set_xticklabels(Labels)
plt.show()
```

```python
#PIE CHART
Labels = ["Man_U", "Man_C", "CHE", "LIV", "ARS", "TOT"]
figure_2, Axes_2 = plt.subplots(figsize = (14,7))
Axes_2.pie(Top_Teams["total_scoring_att"], labels = Labels, autopct = "%1.1f%%", explode = [0.1, 0, 0.1, 0, 0.1, 0])
Axes_2.legend()
plt.show()
```
